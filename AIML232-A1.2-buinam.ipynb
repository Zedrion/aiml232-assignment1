```python
import numpy as np
import matplotlib.pyplot as plt

dimensions = range(1, 21)
fractions = []

for d in dimensions:
    x = np.random.uniform(-1, 1, (10000, d))
    norms = np.linalg.norm(x, axis=1)
    fraction_inside = np.sum(norms < 1) / len(norms)
    fractions.append(fraction_inside)

plt.plot(dimensions, fractions, marker='o')
plt.xlabel("Dimension (d)")
plt.ylabel("Fraction inside unit sphere")
plt.xticks(dimensions)
plt.title("Fraction of Uniform Vectors Inside Unit Sphere vs Dimension", fontsize=14, pad=14)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm
from math import acos
from itertools import combinations

dimensions = [2, 50, 1000]
titles = ["2 Dimensions", "50 Dimensions", "1000 Dimensions"]

fig, axes = plt.subplots(1, 3, figsize=(16,6), sharey=True)
fig.suptitle("Distribution of Angles Between Pairs of Vectors in Different Dimensions", fontsize=20)

for ax, d, title in zip(axes, dimensions, titles):
    x = np.random.normal(0, 1, (50, d))
    pairs = combinations(x, 2)
    angles = [acos(np.dot(a, b) / (norm(a)*norm(b))) for a, b in pairs]
    
    ax.hist(angles, density=True, alpha=0.7, color='blue')
    ax.set_xlim(0, 3)
    ax.set_title(title)
    ax.set_xlabel("Angle (radians)")
    ax.set_ylabel("Density")
    ax.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

# 2.3

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm
from math import acos
from itertools import combinations

dimensions = [2, 50, 1000]
titles = ["2 Dimensions", "50 Dimensions", "1000 Dimensions"]

fig, axes = plt.subplots(1, 3, figsize=(16,6), sharey=True)
fig.suptitle("Distribution of Distances Between Pairs of Vectors in Different Dimensions", fontsize=20)

for ax, d, title in zip(axes, dimensions, titles):
    x = np.random.normal(0, 1, (50, d))
    pairs = combinations(x, 2)
    distances = [norm(a-b) for a, b in pairs]
    
    ax.hist(distances, density=True, alpha=0.7, color='blue')
    ax.set_xlim(0, 50)
    ax.set_title(title)
    ax.set_xlabel("Euclidean Distance")
    ax.set_ylabel("Density")
    ax.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```
